```python
#필요 라이브러리 임포트

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#파일 로딩

df_math = pd.read_csv('/content/drive/MyDrive/빅데이터분석수행/술먹고 공부하기/studywithalcohol/Maths.csv')
df_portuguese = pd.read_csv('/content/drive/MyDrive/빅데이터분석수행/술먹고 공부하기/studywithalcohol/Portuguese.csv')
```

```python
#필요한 자료 추출

df_math_alcohol = df_math[['Dalc', 'Walc']] #평일(Dalc)과 주말(Walc)의 알코올 소비 정도(1~5, 높을수록 많이 소비)
df_math_G3 = df_math['G3'] #G3 점수(1~20, 높을수록 성적 좋음)

df_portuguese_alcohol = df_portuguese[['Dalc', 'Walc']]
df_portuguese_G3 = df_portuguese['G3']
```

```python
def just_avg(series):
  return (series[0] + series[1])/2 #단순 평균

def weight_avg(series):
  return series[0]*(5/7) + series[1]*(2/7) #가중 평균
```

```python
#두 평균 계산 및 dataframe에 추가

m_ja_list = []
m_wa_list = []
p_ja_list = []
p_wa_list = []

for i in range(395):
  value = df_math_alcohol.loc[i]
  m_ja_list.append(just_avg(value))
  m_wa_list.append(weight_avg(value))

df_math_alcohol['weighted_avg'] = m_wa_list
df_math_alcohol['macro_avg'] = m_ja_list

for i in range(649):
  value = df_portuguese_alcohol.loc[i]
  p_ja_list.append(just_avg(value))
  p_wa_list.append(weight_avg(value))

df_portuguese_alcohol['weighted_avg'] = p_wa_list
df_portuguese_alcohol['macro_avg'] = p_ja_list
```

```python
fig = plt.figure()
ax = fig.add_subplot()

x = df_math_alcohol['Dalc'].to_numpy()
y = df_math_G3

ax.set_xlabel('alcohol consumption\n(Dalc)')
ax.set_ylabel('G3 score')

scatter = ax.scatter(x, y, c='r', alpha=0.3, edgecolors='none')

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()

x = df_math_alcohol['Walc'].to_numpy()
y = df_math_G3

ax.set_xlabel('alcohol consumption\n(Walc)')
ax.set_ylabel('G3 score')

scatter = ax.scatter(x, y, c='r', alpha=0.3, edgecolors='none')

plt.show()
```
